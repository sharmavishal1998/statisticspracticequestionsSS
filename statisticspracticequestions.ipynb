````python
#Question1-Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.After generating the list, find the following:
#Write a Python function to calculate the mean of a given list of numbers.Create a function to find the median of a list of numbers.
# Develop a program to compute the mode of a list of integers.
#Implement a function to calculate the weighted mean of a list of values and their corresponding weights.
#Write a Python function to find the geometric mean of a list of positive numbers.
# Create a program to calculate the harmonic mean of a list of values.
#Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).
#Implement a Python program to find the trimmed mean of a list, excluding a certain percentage ofoutliers.
#Answer-
import random
import math
from collections import Counter

# Generate a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# (i): Mean
def calculate_mean(numbers):
    """Calculate the mean of a list of numbers."""
    return sum(numbers) / len(numbers) if numbers else 0

#  (i): Median
def calculate_median(numbers):
    """Calculate the median of a list of numbers."""
    if not numbers:
        return None
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2
    return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2 if n % 2 == 0 else sorted_numbers[mid]

# (ii): Mode
def calculate_mode(numbers):
    """Calculate the mode of a list of numbers."""
    if not numbers:
        return None
    count = Counter(numbers)
    max_frequency = max(count.values())
    mode = [num for num, freq in count.items() if freq == max_frequency]
    return mode if len(mode) > 1 else mode[0]

# (iii): Weighted Mean
def calculate_weighted_mean(values, weights):
    """Calculate the weighted mean of a list of values and weights."""
    if not values or not weights or len(values) != len(weights):
        raise ValueError("Values and weights must be non-empty and of the same length.")
    total_weighted_sum = sum(v * w for v, w in zip(values, weights))
    total_weight = sum(weights)
    return total_weighted_sum / total_weight if total_weight != 0 else 0

#  (iv): Geometric Mean
def calculate_geometric_mean(numbers):
    """Calculate the geometric mean of a list of positive numbers."""
    if not numbers or any(n <= 0 for n in numbers):
        raise ValueError("All numbers must be positive.")
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

# (v): Harmonic Mean
def calculate_harmonic_mean(numbers):
    """Calculate the harmonic mean of a list of numbers."""
    if not numbers or any(n <= 0 for n in numbers):
        raise ValueError("All numbers must be positive.")
    return len(numbers) / sum(1 / num for num in numbers)

# (vi): Midrange
def calculate_midrange(numbers):
    """Calculate the midrange of a list of numbers."""
    if not numbers:
        raise ValueError("The list of numbers cannot be empty.")
    return (min(numbers) + max(numbers)) / 2

# (vii): Trimmed Mean
def calculate_trimmed_mean(numbers, trim_percent):
    """Calculate the trimmed mean by excluding a percentage of outliers."""
    if not numbers or not (0 <= trim_percent < 50):
        raise ValueError("Invalid input or trim percentage.")
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    trim_count = int(n * (trim_percent / 100))
    trimmed_numbers = sorted_numbers[trim_count: n - trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers) if trimmed_numbers else 0

# Example usage:
if __name__ == "__main__":
    print("Generated List:", int_list)
    print("Mean:", calculate_mean(int_list))
    print("Median:", calculate_median(int_list))
    print("Mode:", calculate_mode(int_list))
    values = int_list[:10]  # Example subset of values
    weights = [random.randint(1, 10) for _ in range(10)]
    print("Weighted Mean:", calculate_weighted_mean(values, weights))
    print("Geometric Mean:", calculate_geometric_mean(int_list))
    print("Harmonic Mean:", calculate_harmonic_mean(int_list))
    print("Midrange:", calculate_midrange(int_list))
    print("Trimmed Mean (10%):", calculate_trimmed_mean(int_list, 10))






````

````python
#Question3-Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.
#Answer-
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        """Initialize the discrete random variable with values and their corresponding probabilities.

        Args:
        values (list): The list of possible values of the random variable.
        probabilities (list): The list of corresponding probabilities for each value.
        """
        if len(values) != len(probabilities):
            raise ValueError("The number of values and probabilities must be the same.")
        if not all(0 <= p <= 1 for p in probabilities) or sum(probabilities) != 1:
            raise ValueError("Probabilities must be between 0 and 1 and sum to 1.")

        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        """Calculate the expected value (mean) of the discrete random variable."""
        return sum(v * p for v, p in zip(self.values, self.probabilities))

    def variance(self):
        """Calculate the variance of the discrete random variable."""
        mean = self.expected_value()
        return sum(p * (v - mean) ** 2 for v, p in zip(self.values, self.probabilities))

# Example usage:
values = [1, 2, 3, 4]
probabilities = [0.1, 0.3, 0.4, 0.2]

random_variable = DiscreteRandomVariable(values, probabilities)

# Calculate expected value and variance
expected_val = random_variable.expected_value()
variance_val = random_variable.variance()

print(f"Expected Value: {expected_val}")
print(f"Variance: {variance_val}")
````

````python
#Question04-Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.
#Answer-
import random

class DieSimulation:
    def __init__(self, sides=6):
        """Initialize the die simulation with a specified number of sides (default is 6)."""
        self.sides = sides
        self.outcomes = list(range(1, sides + 1))

    def roll(self):
        """Simulate a single roll of the die and return the outcome."""
        return random.choice(self.outcomes)

    def simulate_rolls(self, num_rolls):
        """Simulate multiple rolls and return a list of outcomes."""
        rolls = [self.roll() for _ in range(num_rolls)]
        return rolls

    def expected_value(self):
        """Calculate the expected value (mean) of a fair six-sided die."""
        return sum(self.outcomes) / len(self.outcomes)

    def variance(self):
        """Calculate the variance of the outcomes of a fair six-sided die."""
        mean = self.expected_value()
        return sum((x - mean) ** 2 for x in self.outcomes) / len(self.outcomes)

# Simulate rolling the die and calculate expected value and variance
die = DieSimulation()

# Simulate rolling the die 1000 times and calculate the expected value and variance
num_rolls = 1000
roll_outcomes = die.simulate_rolls(num_rolls)

# Calculate the expected value and variance
expected_val = die.expected_value()
variance_val = die.variance()

print(f"Expected Value of a fair six-sided die: {expected_val}")
print(f"Variance of a fair six-sided die: {variance_val}")

# For demonstration, show the first 10 outcomes of the simulation
print("\nFirst 10 outcomes of the simulated rolls:", roll_outcomes[:10])

````

````python
#Question5-Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.
#Answer-
import numpy as np

def generate_random_samples(distribution, *params, num_samples=1000):
    """
    Generates random samples from a specified probability distribution and calculates their mean and variance.

    Args:
    distribution (str): The name of the distribution ('binomial' or 'poisson').
    *params: Parameters specific to the distribution.
    num_samples (int): The number of random samples to generate.

    Returns:
    tuple: Mean and variance of the generated samples.
    """
    if distribution == 'binomial':
        n, p = params  # n = number of trials, p = probability of success
        samples = np.random.binomial(n, p, num_samples)
    elif distribution == 'poisson':
        lam = params[0]  # lambda = expected number of occurrences
        samples = np.random.poisson(lam, num_samples)
    else:
        raise ValueError("Unsupported distribution. Use 'binomial' or 'poisson'.")

    # Calculate mean and variance of the generated samples
    mean = np.mean(samples)
    variance = np.var(samples)

    return mean, variance

# Example usage:

# Binomial distribution example (n=10 trials, p=0.5 probability of success)
binomial_mean, binomial_variance = generate_random_samples('binomial', 10, 0.5)
print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")

# Poisson distribution example (lambda = 5 expected occurrences)
poisson_mean, poisson_variance = generate_random_samples('poisson', 5)
print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")
````

````python
#Question6-Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples
#Answer-
import numpy as np

def generate_gaussian_samples(mean=0, std_dev=1, num_samples=1000):
    """
    Generate random samples from a Gaussian (normal) distribution and calculate their mean, variance, and standard deviation.

    Args:
    mean (float): The mean (mu) of the normal distribution.
    std_dev (float): The standard deviation (sigma) of the normal distribution.
    num_samples (int): The number of samples to generate.

    Returns:
    tuple: Mean, variance, and standard deviation of the generated samples.
    """
    # Generate random samples from a normal distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

    # Calculate mean, variance, and standard deviation
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)

    return sample_mean, sample_variance, sample_std_dev

# Example usage:

# Gaussian distribution with mean=0 and standard deviation=1
mean, variance, std_dev = generate_gaussian_samples(mean=0, std_dev=1, num_samples=1000)

print(f"Generated Gaussian Distribution - Mean: {mean}, Variance: {variance}, Standard Deviation: {std_dev}")
````

````python
#Question08- Write a Python function to calculate the probability density function (PDF) of a continuous random variable for a given normal distribution.
#Answer-
import numpy as np

def normal_pdf(x, mean=0, std_dev=1):
    """
    Calculate the Probability Density Function (PDF) of a normal distribution.

    Args:
    x (float): The point at which to calculate the PDF.
    mean (float): The mean (mu) of the normal distribution (default is 0).
    std_dev (float): The standard deviation (sigma) of the normal distribution (default is 1).

    Returns:
    float: The value of the PDF at point x.
    """
    # Calculate the PDF using the normal distribution formula
    pdf_value = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    return pdf_value

# Example usage:
x = 2  # Point where we want to calculate the PDF
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution

pdf_value = normal_pdf(x, mean, std_dev)
print(f"The PDF value at x={x} for N({mean}, {std_dev}^2) is {pdf_value}")

````

````python
#QUESTION9-Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.
#Answer-
import numpy as np

def exponential_cdf(x, lambda_param):
    """
    Calculate the Cumulative Distribution Function (CDF) of an exponential distribution.

    Args:
    x (float): The point at which to calculate the CDF.
    lambda_param (float): The rate parameter (lambda) of the exponential distribution.

    Returns:
    float: The value of the CDF at point x.
    """
    # Ensure x is non-negative for exponential distribution
    if x < 0:
        raise ValueError("x must be a non-negative value for the exponential distribution.")

    # Calculate the CDF using the formula: F(x; λ) = 1 - exp(-λx)
    cdf_value = 1 - np.exp(-lambda_param * x)
    return cdf_value

# Example usage:
x = 3  # Point at which to calculate the CDF
lambda_param = 0.5  # Rate parameter (λ)

cdf_value = exponential_cdf(x, lambda_param)
print(f"The CDF value at x={x} for Exponential(λ={lambda_param}) is {cdf_value}")
````

````python
#Question10- Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.
#Answer-
import math

def poisson_pmf(k, lambda_param):
    """
    Calculate the Probability Mass Function (PMF) of a Poisson distribution.

    Args:
    k (int): The value at which to calculate the PMF (must be a non-negative integer).
    lambda_param (float): The rate parameter (λ) of the Poisson distribution.

    Returns:
    float: The PMF value at k.
    """
    if k < 0 or not isinstance(k, int):
        raise ValueError("k must be a non-negative integer.")

    # Calculate the Poisson PMF: P(X = k) = (λ^k * exp(-λ)) / k!
    pmf_value = (lambda_param ** k) * math.exp(-lambda_param) / math.factorial(k)
    return pmf_value

# Example usage:
k = 3  # Value at which to calculate the PMF
lambda_param = 2.5  # Rate parameter (λ)

pmf_value = poisson_pmf(k, lambda_param)
print(f"The PMF value at k={k} for Poisson(λ={lambda_param}) is {pmf_value}")
````

````python
#Question11-A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
# who make a purchase). They collect data from the old and new layouts to compare.
# To generate the data use the following command:
# ```python
# import numpy as np
# # 50 purchases out of 1000 visitors
# old_layout = np.array([1] * 50 + [0] * 950)
# # 70 purchases out of 1000 visitors
# new_layout = np.array([1] * 70 + [0] * 930)
# Apply z-test to find which layout is successful.

#Answer-
import numpy as np
import scipy.stats as stats

# Given data
old_layout = np.array([1] * 50 + [0] * 950)  # 50 purchases out of 1000 visitors
new_layout = np.array([1] * 70 + [0] * 930)  # 70 purchases out of 1000 visitors

# Calculate sample proportions
p1_hat = np.mean(old_layout)  # Proportion of purchases for old layout
p2_hat = np.mean(new_layout)  # Proportion of purchases for new layout

# Sample sizes
n1 = len(old_layout)
n2 = len(new_layout)

# Pooled proportion
p_hat = (np.sum(old_layout) + np.sum(new_layout)) / (n1 + n2)

# Standard error
SE = np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))

# Z-score
z = (p1_hat - p2_hat) / SE

# Calculate p-value (one-tailed test, since we're testing if the new layout is better)
p_value = 1 - stats.norm.cdf(z)

# Print results
print(f"Old layout conversion rate: {p1_hat:.4f}")
print(f"New layout conversion rate: {p2_hat:.4f}")
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")
````

````python
# #Question12-A tutoring service claims that its program improves students' exam scores. A sample of students who
# participated in the program was taken, and their scores before and after the program were recorded.
# Use the below code to generate samples of respective arrays of marks:
# ```python
# before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
# after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])
# ```
# Use z-test to find if the claims made by tutor are true or false

#Answer-
import numpy as np
import scipy.stats as stats

# Sample data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Step 1: Calculate the differences (after - before)
differences = after_program - before_program

# Step 2: Compute the sample mean and sample standard deviation of the differences
mean_diff = np.mean(differences)
std_diff = np.std(differences, ddof=1)

# Step 3: Calculate the standard error (SE) of the differences
n = len(differences)
SE = std_diff / np.sqrt(n)

# Step 4: Calculate the z-score
z_score = mean_diff / SE

# Step 5: Calculate the p-value for one-tailed test (since we expect improvement)
p_value = 1 - stats.norm.cdf(z_score)

# Print the results
print(f"Mean of differences: {mean_diff:.4f}")
print(f"Standard deviation of differences: {std_diff:.4f}")
print(f"Standard error: {SE:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision based on p-value (significance level of 0.05)
if p_value < 0.05:
    print("Reject the null hypothesis: The tutoring program has significantly improved students' scores.")
else:
    print("Fail to reject the null hypothesis: There is no significant improvement in students' scores.")
````

````python
# #Question15-A company is testing two different website layouts to see which one leads to higher click-through rates.
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
# freedom, and p-value.
# Use the following data:
# ```python
# layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
# layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

#Answer-
import numpy as np
from scipy import stats

# Sample data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Step 1: Calculate the sample means and sample standard deviations
mean_a = np.mean(layout_a_clicks)
mean_b = np.mean(layout_b_clicks)
std_a = np.std(layout_a_clicks, ddof=1)
std_b = np.std(layout_b_clicks, ddof=1)

# Step 2: Calculate the sample sizes
n_a = len(layout_a_clicks)
n_b = len(layout_b_clicks)

# Step 3: Calculate the pooled standard deviation
pooled_sd = np.sqrt(((std_a**2) / n_a) + ((std_b**2) / n_b))

# Step 4: Calculate the t-statistic
t_stat = (mean_a - mean_b) / pooled_sd

# Step 5: Calculate the degrees of freedom
df = ((std_a**2 / n_a + std_b**2 / n_b)**2) / (((std_a**2 / n_a)**2 / (n_a - 1)) + ((std_b**2 / n_b)**2 / (n_b - 1)))

# Step 6: Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Print the results
print(f"Mean of Layout A: {mean_a:.4f}")
print(f"Mean of Layout B: {mean_b:.4f}")
print(f"Standard Deviation of Layout A: {std_a:.4f}")
print(f"Standard Deviation of Layout B: {std_b:.4f}")
print(f"Pooled Standard Deviation: {pooled_sd:.4f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"Degrees of Freedom: {df:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision based on p-value (significance level of 0.05)
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference between the click-through rates of the two layouts.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the click-through rates of the two layouts.")


````

````python
# #Question17- A school district introduces an educational intervention program to improve math scores. Write a Python
# function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
# determine if the intervention had a significant impact.
# Use the following data of test score:
# ```python
# pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
# post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
#answer-
import numpy as np
from scipy import stats

def analyze_intervention(pre_scores, post_scores):
    # Step 1: Calculate the differences between paired scores
    differences = np.array(post_scores) - np.array(pre_scores)

    # Step 2: Calculate the mean and standard deviation of the differences
    mean_diff = np.mean(differences)
    std_diff = np.std(differences, ddof=1)

    # Step 3: Calculate the t-statistic
    n = len(differences)  # Number of paired samples
    t_stat = mean_diff / (std_diff / np.sqrt(n))

    # Step 4: Calculate the degrees of freedom
    df = n - 1

    # Step 5: Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df))  # Multiply by 2 for two-tailed test

    # Print the results
    print(f"Mean of differences: {mean_diff:.4f}")
    print(f"Standard deviation of differences: {std_diff:.4f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of Freedom: {df}")
    print(f"P-value: {p_value:.4f}")

    # Step 6: Decision based on p-value (significance level of 0.05)
    if p_value < 0.05:
        print("Reject the null hypothesis: The intervention had a significant impact on the test scores.")
    else:
        print("Fail to reject the null hypothesis: The intervention did not have a significant impact on the test scores.")

# Data for pre- and post-intervention test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Call the function to analyze the intervention impact
analyze_intervention(pre_intervention_scores, post_intervention_scores)
````

````python
# #Question18-An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
# a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
# significant difference between the average salaries of male and female employees.
# Use the below code to generate synthetic data:
# ```python
# # Generate synthetic salary data for male and female employees
# np.random.seed(0)  # For reproducibility
# male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
# female_salaries = np.random.normal(loc=55000, scale=9000, size=20)
#Answer-
import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

# Perform independent two-sample t-test
def analyze_salary_gap(male_salaries, female_salaries):
    # Step 1: Calculate the sample means and standard deviations
    mean_male = np.mean(male_salaries)
    std_male = np.std(male_salaries, ddof=1)

    mean_female = np.mean(female_salaries)
    std_female = np.std(female_salaries, ddof=1)

    # Step 2: Calculate the t-statistic
    n_male = len(male_salaries)
    n_female = len(female_salaries)

    t_stat = (mean_male - mean_female) / np.sqrt((std_male**2 / n_male) + (std_female**2 / n_female))

    # Step 3: Calculate the degrees of freedom
    df = ((std_male**2 / n_male) + (std_female**2 / n_female))**2 / \
         (((std_male**2 / n_male)**2 / (n_male - 1)) + ((std_female**2 / n_female)**2 / (n_female - 1)))

    # Step 4: Calculate the p-value
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_stat), df))  # Two-tailed test

    # Print the results
    print(f"Mean male salary: {mean_male:.2f}")
    print(f"Standard deviation of male salary: {std_male:.2f}")
    print(f"Mean female salary: {mean_female:.2f}")
    print(f"Standard deviation of female salary: {std_female:.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"Degrees of freedom: {df:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Step 5: Decision based on p-value (significance level of 0.05)
    if p_value < 0.05:
        print("Reject the null hypothesis: There is a significant salary gap between male and female employees.")
    else:
        print("Fail to reject the null hypothesis: There is no significant salary gap between male and female employees.")

# Analyze salary gap
analyze_salary_gap(male_salaries, female_salaries)
````

````python
#Question21-A political analyst wants to determine if there is a significant association between age groups and voter
# preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
# them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
# there is a significant association between age groups and voter preferences.
# Use the below code to generate data:
# ```python
# np.random.seed(0)
# age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)
# voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

#Answer-
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate synthetic data using the provided code
np.random.seed(0)

age_groups = np.random.choice(['18-30', '31-50', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

# Create a DataFrame from the generated data
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Step 1: Create the contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Step 2: Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Step 3: Display the results
print("Contingency Table:")
print(contingency_table)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
print("P-Value:", p_value)

# Step 4: Decision based on p-value
if p_value < 0.05:
    print("\nReject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("\nFail to reject the null hypothesis: There is no significant association between age groups and voter preferences.")
````

````python
# #Question22-A company conducted a customer satisfaction survey to determine if there is a significant relationship
# between product satisfaction levels (Satisfied, Neutral, Dissatisfied) and the region where customers are
# located (East, West, North, South). The survey data is summarized in a contingency table. Conduct a ChiSquare test to determine if there is a significant relationship between product satisfaction levels and
# customer regions.
# Sample data:
# ```python
# #Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
# data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

#Answer-
import numpy as np
from scipy.stats import chi2_contingency

# Sample data: Product satisfaction levels (rows) vs. Customer regions (columns)
data = np.array([[50, 30, 40, 20], [30, 40, 30, 50], [20, 30, 40, 30]])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Display the results
print("Contingency Table:")
print(data)
print("\nChi-Square Statistic:", chi2_stat)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected)
print("P-Value:", p_value)

# Decision based on p-value
if p_value < 0.05:
    print("\nReject the null hypothesis: There is a significant relationship between product satisfaction levels and customer regions.")
else:
    print("\nFail to reject the null hypothesis: There is no significant relationship between product satisfaction levels and customer regions.")
````

````python
# #Question24-A company produces three different versions of a product: Standard, Premium, and Deluxe. The
# company wants to determine if there is a significant difference in customer satisfaction scores among the
# three product versions. They conducted a survey and collected customer satisfaction scores for each
# version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
# difference in customer satisfaction scores.
# Use the following data:
# ```python
# # Sample data: Customer satisfaction scores for each product version
# standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
# premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
# deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

#ANSWER-
import numpy as np
from scipy.stats import f_oneway

# Sample data: Customer satisfaction scores for each product version
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

# Perform the ANOVA test
f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

# Display the results
print("ANOVA Test Results:")
print("F-Statistic:", f_stat)
print("P-Value:", p_value)

# Decision based on p-value
if p_value < 0.05:
    print("\nReject the null hypothesis: There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference in customer satisfaction scores among the product versions.")
````
